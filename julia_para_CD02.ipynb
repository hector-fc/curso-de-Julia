## Condicionais 
Condicionais servem para controlar o fluxo do programa. Em outras palavras, partes do código só serão executadas se determinada condição, de valor booleano, for verdadeira. Vejamos um exemplo

````julia
expr = true
if expr 
    println("Ola ! ")
end
````

````julia
expr1 = true
expr2 = false
if expr1 
    println("Expr 1 é verdadeira ! ")
    println("Expr 2 é false ! ")
end
if expr2 
    println("Expr 1  não será impressa ! ")
    println("Expr 2   não será impresa! ")
end
````

````julia
expr1 = true
expr2 = false
if expr1 
    println("Expr 1 é verdadeira ! ")
    println("Expr será impresa ! ")
elseif expr2 
    println("Expr 1  não será impressa ! ")
    println("Expr 2   não será impresa! ")
end
````

## Repitição FOR
Algumas vezes, uma mesma sequência de comandos deve ser executada diversas vezes. Para fazer isso utilizamos estruturas de repetição. Observe o código:

````julia
x = 1 
x = x + 2
x = x + 2 
x = x + 2 
x = x + 2 
println(x)
````

Como as linha 2, 3, 4 e 5, são repetidas, podemos utilizar um estrutura de repetição. O for é a keyword utilizada para tal. Vejamos:

````julia
x = 1 
for i in [1,2,3,4]
    x = x + 2     
end
println(x)
````

````julia
x = 1 
for i in [1,2,3,4]
    x = x + 2     
    println("i= ",i,", x= ",x)
end
println(x)
````

## Funções 
Funções são utilizadas para agrupar uma série de comandos na intenção de reutilizálos. Por exemplo, se você quer saber se três variáveis X, Y e Z têm valores inteiros pares, uma opção seria:

````julia
X = 7
if mod(X,2) == 0
    println("Par")
else
    print("Impar")
end

Y = 2 
if mod(Y,2) == 0
    print("Par")
else
    print("Impar")
end

Z = 42
if mod(Z,2) == 0
    print("Par")
else
    print("Impar")
end
````

````julia
function isPar(var)
    if mod(var,2)==0
        println("Par")
    else
        println("impar")
    end
end
````

````julia
isPar(X)
isPar(Y)
isPar(Z)
````

## Função Fatorial 
Esta seção tem o objetivo de praticar quase tudo visto nesse tutorial. Para isso criaremos uma função, chamada fatorial, que retorna o fatorial de um número n dado como parâmetro de entrada da função.

````julia
function fatorial(n)
    if n == 0
        return 1 
    elseif n==1
        return 1 
    else
        resultado = 1 
        for i in 1:n
            resultado=resultado*i
        end    
        return resultado        
    end        
end    
    
````

````julia
fatorial(0)
````

````julia
fatorial(5)
````

### Recursiva

````julia
function fatorial(n)
    if n == 0
        return 1 
    elseif n==1
        return 1 
    else
       n*fatorial(n-1)
    end        
end    
````

````julia
fatorial(5)
````

# Divisão Euclidiana 

----------------
**Teorema** 
Dados $a$ e $b$ positivos, com $a>b$  então existe $q,r\in\mathbb{Z}$ tal que  $a = bq+r$ e 
$mdc(a,b)=mdc(b,r)$
----------------




## Algoritmo de euclides 
Dados $a,b$ com $a>b$ 
1. Calcular $q,r\in \mathbb{Z}$  tal que  $a = bq+r$.
2. Se $r=0$, parar. mdc(a,b)=b 
3. Fazer $a=b, b=r$ e voltar ao primer passo 1




````julia
using Printf

a = 322
b = 72

q = Int(floor(a/b)) 
r = a - b*q
@printf("%5i = %5i x %5i + %5i\n",a,b,q,r)
a,b=b,r 
````

````julia
using Printf
#(32,12)

a = 32
b = 12
while true 
    q = Int(floor(a/b)) 
    r = a - b*q
    @printf("%5i = %5i x %5i + %5i\n",a,b,q,r)
    if r==0
        @printf("O máximo divisor comum é: %5i",b)
        break
    end
    a,b=b,r 
end
````

````julia
using Printf

function mdc1(a,b) 
    while true 
        q = Int(floor(a/b)) 
        r = a - b*q        
        if r==0                        
            return b 
        end
        a,b=b,r 
    end
    return b
end 
mdc1(32,12)    
````

````julia
function mdc1(a,b)     
    q = Int(floor(a/b)) 
    r = a - b*q        
    if (r == 0)  return b else  mdc1(b,r)  end       
end 

@time mdc1(10030,502)    
````

````julia
function phiE(n)
    sum = 0 
    for i in 1:n
        if mdc2(i,n)== 1
           sum+=1 
        end 
    end    
    return sum 
end

phiE(17)
````
