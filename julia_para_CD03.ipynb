# VizualiÃ§Ã£o 

```julia
using Plots
```

```julia
theme(:dark)
```

```julia
X = iris.PetalLength
Y = iris.SepalLength

labels=iris.Species

scatter(X,
    Y,
    xlim =(0,7.5),
    ylim =(3.5,8.5),
    xlabel="petal length (Comprimento)",
    ylabel="sepal length (Comprimento)",
    title= "Iris Dataset",
    group =labels,
    legend = :topleft,
    color=[:darkviolet :green3 :red3],
    markersize=[3 5 3],
    alpha = 0.9,
    markershape =[:hex :star5 :o],
    dpi = 500
)
```
